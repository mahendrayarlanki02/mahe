# Treat Outliers

```python
continous=["Unit price","Quantity","Tax 5%","Sales","cogs", "gross income","Rating"]
categorical=["Invoice ID", "Branch"," City"," Customer type"," Gender"," Product line"," Date"," Time"," Payment"]
```

```python
for i in continous:
    sns.boxplot(df[i])
    plt.show()
```

```python
for i in continous:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - (1.5 * IQR)
    upper_limit = Q3 + (1.5 * IQR)
    
    outliers = df[(df[i] < lower_limit) | (df[i] > upper_limit)]
    print(f"Outliers in {i}:")
    print(outliers[[i]].reset_index(drop=True)) 
```

### 10.Which Product line has the highest customer Ratings?

### 11.What is the trend of daily sales over time?

```python
df["Date"] = pd.to_datetime(df["Date"])
daily_sales = df.groupby("Date")["Sales"].sum()
daily_sales.plot(figsize=(10,5), title="Daily Sales Trend")
plt.show()
```

### 12.Which day of the week has the highest sales?

```python
df["DayOfWeek"] = df["Date"].dt.day_name()
sns.barplot(x="DayOfWeek", y="Sales", data=df, estimator=sum,
            order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
plt.title("Sales by Day of Week")
plt.show()
```

### 13.At what time of day (morning, afternoon, evening) are most purchases made?

```python
def time_of_day(x):
    hour = int(x.split(":")[0])
    if 6 <= hour < 12: return "Morning"
    elif 12 <= hour < 17: return "Afternoon"
    else: return "Evening"
df["TimeOfDay"] = df["Time"].apply(time_of_day)

sns.barplot(x="TimeOfDay", y="Sales", data=df, estimator=sum)
plt.title("Sales by Time of Day")
plt.show()
```

### 14.Do Ratings vary by time of purchase?

```python
sns.boxplot(x="TimeOfDay", y="Rating", data=df)
plt.title("Customer Ratings by Time of Day")
plt.show()
```

### 15.Which month has the highest sales?

```python
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales.plot(kind="bar", title="Monthly Sales")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["Quantity", "Rating", "Sales", "Unit price"]].corr(), 
            annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

### 31.Which product line contributes the most to total revenue in each branch?

```python
branch_revenue = df.groupby(["Branch","Product line"])["Sales"].sum().reset_index()
pivot = branch_revenue.pivot(index="Product line", columns="Branch", values="Sales")
pivot.plot(kind="bar", figsize=(10,5))
plt.title("Revenue Contribution by Product Line per Branch")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

# Conclusion

### This analysis explored supermarket transactional data to understand revenue drivers, customer behavior, and operational performance across branches. The results indicate that sales performance is influenced more by product category demand and customer purchasing patterns than by location alone. Certain product lines consistently contribute higher revenue, while others generate volume without proportional profitability.
### Customer type and payment preferences also play a significant role in transaction frequency, with returning customers contributing more stable revenue streams. Additionally, peak shopping periods occur during specific days and hours, suggesting opportunities for optimized staffing and promotional planning.
### Overall, the findings demonstrate that targeted promotions, inventory planning, and customer-focused strategies can improve revenue consistency and operational efficiency. Data-driven decision making can help supermarkets reduce idle inventory, enhance customer satisfaction, and improve profitability.
