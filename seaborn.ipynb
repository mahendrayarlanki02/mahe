 ->PLOTS  FOR CONTINOUS DATA
-----------------------------

1. univariate (single variable)
- histogarm
- boxplot
2. bivariate(plots between two variables)
- scatter plot
- line plot
- joint plot
- violin plot

3.multivariate (more than 2 variables)
- pairplot
- heatmap

```python
# histogram/ distribution plot
sns.displot(df["total_bill"],bins=10,kde=True)
plt.show()
```

```python
# box plot
sns.boxplot(y=df["tip"])
plt.show()
```

```python
# scattter plot
sns.scatterplot(x=df["total_bill"],y=df["tip"])
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"])
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],data=df,hue="sex")
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],data=df,style="sex")
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],data=df,size="sex")
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],data=df,hue="sex",style="smoker")
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],data=df,col="size")
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],data=df,hue="sex",col="size",col_wrap= 2)
plt.show()
```

```python
# lineplot
sns.relplot(x="total_bill",y="tip",data=df,kind='line')
plt.show()
```

```python
df["sno"]=pd.DataFrame(np.arange(1,245))
df
```

```python
sns.relplot(x="sno",y="tip",kind='line',data=df)
plt.show()
```

```python
df.drop("sno",axis=1,inplace=True)
```

```python
# joint plot
sns.jointplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
# violin plot
sns.violinplot(x="total_bill",y="day",data=df)
plt.show()
```

```python
# pair plot
sns.pairplot(df,vars=continous)
plt.show()
```

```python
# heat map
mc=df[["total_bill","tip"]].corr()
mc
```

```python
sns.heatmap(mc,annot=True)
plt.show()
```

**->PLOTS FOR DISCRETE DATA**
-----------------------------

1.univariate (single variable)
- pie plot

2.bivariate(plots between two variables)
- box plot (1D+1C)


```python
# count plot
df["smoker"].value_counts()
```

```python
sns.countplot(x=df["smoker"])
plt.show()
```

```python
sns.countplot(x=df["smoker"])
plt.show()
```

```python
# barplot
sns.catplot(x="sex",y="total_bill",data=df)
plt.show()
```

```python
sns.catplot(x="sex",y="total_bill",data=df,kind='bar')
plt.show()
```

```python
sns.catplot(x="day", y  ="total_bill", data=df ,kind="bar", hue="size")
plt.show()
```

```python
sns.catplot(x="smoker", y  ="tip", data=df ,kind="bar", hue="size")
plt.show()
```

```python
# boxplot
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.catplot(x="day", y  ="total_bill", hue="smoker",data=df,kind="box")
plt.show()
```
