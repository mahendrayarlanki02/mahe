# Treat outliers

```python
continous=["Billing Amount","Age","Room Number"]
categorical=["Name","Gender","Blood Type","Medical Condition","Doctor","Insurance Provider","Admission Type","Medication","Test Results"]
```

```python
for i in continous:
    sns.boxplot(df[i])
    plt.show()
```

```python
for i in continous:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - (1.5 * IQR)
    upper_limit = Q3 + (1.5 * IQR)
    
    outliers = df[(df[i] < lower_limit) | (df[i] > upper_limit)]
    print(f"Outliers in {i}:")
    print(outliers[[i]].reset_index(drop=True)) 
```

## 9.What is the trend of patient visits across months or years (if a date column exists)?

```python
# Convert Date column to datetime (assuming column name is 'Date')
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'], errors='coerce')

# Extract Year and Month
df['Year'] = df['Date of Admission'].dt.year
df['Month'] = df['Date of Admission'].dt.month

# Count visits per month
monthly_visits = df.groupby(['Year', 'Month']).size().reset_index(name='Visits')

# Create a datetime column for plotting
monthly_visits['YearMonth'] = pd.to_datetime(monthly_visits['Year'].astype(str) + '-' + monthly_visits['Month'].astype(str) + '-01')

# Visualization
plt.figure(figsize=(10,5))
sns.lineplot(x='YearMonth', y='Visits', data=monthly_visits, marker="o")
plt.title("Patient Visit Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Visits")
plt.xticks(rotation=45)
plt.show()
```

## 10.Create a pivot table of Gender vs. Disease Outcome, and compute survival/recovery rates."

```python

pivot = pd.crosstab(df['Gender'], df['Test Results'], normalize='index') * 100
print("Disease Outcome by Gender (%):\n", pivot.round(2))

# Visualization (stacked bar chart)
pivot.plot(kind='bar', stacked=True, figsize=(8,5), colormap="viridis")
plt.title("Disease Outcome Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Percentage (%)")
plt.legend(title="Disease Outcome")
plt.show()
```

Implementing early screening based on these indicators can reduce hospital load and improve patient outcomes. Overall, data-driven monitoring can support better healthcare planning and resource allocation.
